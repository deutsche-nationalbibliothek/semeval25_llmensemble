```python
# Functions helping to determine, how many different combinations of promptsxmodels are possible, when assuming different numbers of combinations
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
def possible_combinations(n,k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# Define the range of n and k
n_values = range(1, 30)
k_values = range(1, 5)

# Print the table header
print(f"{'n':<5}{'k':<5}{'Combinations':<15}")

# Iterate over n and k values and print the outcomes
for n in n_values:
    for k in k_values:
        if k <= n:
            print(f"{n:<5}{k:<5}{possible_combinations(n, k):<15}")
```

```python
# Define the range of n and k
n_values = range(1, 30)
k_values = range(1, 7)

# Print the table header
print(f"{'n/k':<5}", end="")
for k in k_values:
    print(f"{k:<15}", end="")
print()

# Iterate over n values and print the outcomes for each k
for n in n_values:
    print(f"{n:<5}", end="")
    for k in k_values:
        if k <= n:
            print(f"{possible_combinations(n, k):<15}", end="")
        else:
            print(f"{'':<15}", end="")
    print()
```

```python
from itertools import combinations

def sample_sublists(lst, n):
    return list(combinations(lst, n))

# Example usage
items = ["mp1", "mp2", "mp3", "mp4", "mp5", "mp6"]
n = 2
sublists = sample_sublists(items, n)
print(sublists)  # Output: [(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (3, 4)]
print(len(sublists))
```
